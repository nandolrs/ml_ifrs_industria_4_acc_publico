```python

def DadoCarregar(data_url):
    import pandas as pd
    from decimal import Decimal

    dados = pd.read_json(data_url) 

    titulos = dados.iloc[0]

    dados_ = dados.copy(deep=True)
    
    #X_data = X_data.drop(columns=['MC','MN','D1C','D2C','D2N','NC','NN','D3C','D3N'])
    
    dados_ = dados_.drop(0, axis=0) 
    
    
    #X_data['V'].describe()
    dados_['V'] = dados_['V'].astype('float64')
    
    dados_ = dados_.sort_values(by='V')
    
    #print(X_data.dtypes)
    
    return dados_, titulos
```

```python

def GraficoGerar(data_categorias, data_valores, titulos_categorias, titulo):
    import matplotlib.pyplot as plt
    #fig, ax = plt.subplots(figsize=(10*0.6, 5*0.95), layout='constrained')
    fig, ax = plt.subplots()
    
    plt.xticks(rotation=90)
    
    #X_data_categorias = X_data['D1N']
    #X_data_valores = X_data['V']
    
    low_x, high_x = ax.get_xlim()
    low_y, high_y = ax.get_ylim()
    
    bar_container = ax.bar(data_categorias, data_valores)  # , align='edge'
    ax.set(xlabel=titulos_categorias, title=titulo) # , ylim=(low_y, high_y) X_titulos['D1N'] X_titulos['D4N']
    #ax.bar_label(bar_container, fmt='{:,.0f}')
```

tabela 4678

```python
data_url = "https://apisidra.ibge.gov.br/values/t/4678/n3/all/v/9819/p/2022/d/v9819%201" 

data, titulos = DadoCarregar(data_url)
GraficoGerar(data_categorias=data['D1N'], data_valores=data['V'], titulos_categorias=data['D2N'], titulo=data.iloc[0]['D2N'])
```
