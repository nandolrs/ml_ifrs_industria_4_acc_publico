```python
import pandas as pd
from decimal import Decimal

url4678 = "https://apisidra.ibge.gov.br/values/t/4678/n3/all/v/9819/p/2022/d/v9819%201" 
X_full4678 = pd.read_json(url4678) 
#X_full4678 = pd.read_xml(url4678, xpath=".//ArrayOfValorDescritoPorSuasDimensoes") 


url9471 = "https://apisidra.ibge.gov.br/values/t/9471/n3/all/v/12965/p/all/c1675/59804/d/v12965%201" 
X_full9471 = pd.read_json(url9471) 
#X_full9471 = pd.read_xml(url9471, xpath=".//ArrayOfValorDescritoPorSuasDimensoes") 


X_titulos = X_full9471.iloc[0]

X_data = X_full9471.copy(deep=True)

X_data = X_data.drop(columns=['MC','MN','D1C','D2C','D2N','NC','NN','D3C','D3N'])

X_data = X_data.drop(0, axis=0) 


#X_data['V'].describe()
X_data['V'] = X_data['V'].astype('float64')

X_data = X_data.sort_values(by='V')

#print(X_data.dtypes)

X_data
```
